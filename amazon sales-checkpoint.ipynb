#SIZE 

```python
ax = sns.countplot(x ='Size',data = df)
ax.bar_label(ax.containers[0])
plt.show()
```

note: from the above graph we know most of the people buy m size 

```python
gb = df.groupby(["Size"]).agg({"Qty": "sum"})

# Plotting the bar chart
plt.bar(gb.index, gb["Qty"])

sorted_gb = gb.sort_values(by='Qty', ascending=False)

# Adding labels and title
plt.xlabel('Size')
plt.ylabel('Sum of Quantity')
plt.title('Sum of Quantity by Size')
plt.show()

print(sorted_gb)
```

 note:from the above data we have seen that M size is brought more oftenly

# histogram 

```python
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize = (10, 5 ))
plt.hist(column_data , bins = 20 , edgecolor = 'Black' ) 
plt.xticks (rotation = 90)
plt.show()
```

note :from the above chart we have noticed that max buyers are of Tshirts 

```python
B2B_check = df['B2B'].value_counts()
plt.pie(B2B_check , labels = B2B_check.index , autopct = '%1.1f%%')
plt.show()
```

note: from the above chart we have noticed that max buyers are retailers i.e about 99.2% and the rest 0.8% are B2B buyers 

```python
#Prepare data fro scatter plot 
x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data , y_data)
plt.xlabel = ('Category')
plt.ylabel = ('Size')
plt.title =  ('Scatter Plot')
plt.show()
```

note :plot count of cities by state 

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')

# Adding labels and title
plt.xlabel('Ship State')  # Correct label for x-axis
plt.ylabel('Count')       # Correct label for y-axis
plt.title('Distribution of Ship States')  # Title of the plot

# Rotating x-ticks for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

note:from the above chart we have noted that maximum buyers are from maharashtra
